# Data Science Project for Ionosphere DataSet 

```python
# import libraries needed in project
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.feature_selection import RFE
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.metrics import classification_report, recall_score, accuracy_score, f1_score, confusion_matrix, precision_score
from sklearn.svm import SVC
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import RobustScaler, PowerTransformer, PolynomialFeatures, OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import learning_curve
import matplotlib.pyplot as plt
from sklearn.metrics import roc_auc_score, roc_curve, auc
import seaborn as sns
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import make_scorer

```

## Pre Processing Data

```python
dataset=pd.read_csv("Path\\ionosphere_data.csv")
X=dataset.iloc[:,0:33]
y=dataset.iloc[:,34]
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size = 0.3, random_state=0)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Logistic Regression Model

```python
def logistic_regression():
    
    #Create a model and print classification report
    LR = LogisticRegression()
    LR.fit(X_train,y_train)
    predictions = LR.predict(X_test)
    print("Logistic Regression Classification Report")
    print(classification_report(y_test, predictions))
    
    # plot Confusion matrix
    cm_LR_test = confusion_matrix(y_test, predictions, labels = ['b','g'])
    df_LR_test = pd.DataFrame(data=cm_LR_test, columns=['Predicted b', 'Predicted g'], index=['Actual b', 'Actual g'])
    sns.heatmap(df_LR_test, annot=True)
    plt.show()
    
def logistic_regression_learningcurve():
    
    #Logistic Regression learning curve for overfitting and underfitting analysis
    pline=make_pipeline(StandardScaler(),LogisticRegression())
    model=pline.fit(X,y)
    train_sizes, train_scores, test_scores=learning_curve(model, X, y,cv=5,train_sizes=np.linspace(.1, 1.0, 5),shuffle=True)
    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)
    plt.plot(train_sizes,train_mean,label="Training Accuracy")
    plt.plot(train_sizes,test_mean,label="Validation Accuracy")
    plt.title("Learning Curve")
    plt.xlabel("Number of data points")
    plt.ylabel("Accuracy")
    plt.legend(loc="best")

def logistic_regression_feature_selection():

    # Classification with Featre Selection RFE
    featureNames=X.columns
    LR=LogisticRegression()
    rfeModel=RFE(LR,n_features_to_select=16)
    rfeSelection=rfeModel.fit(X,y)
    rfeSelection.ranking_
    rfeSelection.support_
    rfeFeatures=np.array(featureNames)[rfeSelection.support_]
    XNew = dataset[rfeFeatures]
    X_train, X_test, y_train, y_test = train_test_split(XNew,y, test_size = 0.3, random_state=0)
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
    LR.fit(X_train,y_train)
    predictions = LR.predict(X_test)
    print("Logistic Regression classification Report with Feature Selection")
    print(classification_report(y_test, predictions))
    
def logistic_regression_auc_roc_curve():
    
    #Logistic Regression model Auc-Roc curve plotting
    pline = Pipeline([('scale', StandardScaler()),('algo', LogisticRegression())])
    pline.fit(X_train, y_train)
    y_proba = pline.predict_proba(X_test)
    pred_1 = y_proba[:, 1]
    roc_curve(y_test, pred_1,pos_label = 'g')
    falsePR, truePR, threshold = roc_curve(y_test, pred_1, pos_label = 'g')
    aucScore = round((auc(falsePR, truePR)), 2)
    plt.figure(figsize=(10, 6))
    plt.plot(falsePR, truePR, 'g', label = f"AUC = {aucScore}")
    plt.plot([0, 1], [0, 1], 'r-.')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.xlabel('False Positive Rate ')
    plt.ylabel('True Positive Rate')
    plt.title('ROC-AUC curve')
    plt.legend()
    plt.show()
    
```

## Logistic Regression Model Results

```python
logistic_regression()
logistic_regression_learningcurve()
logistic_regression_feature_selection()
logistic_regression_auc_roc_curve()
```

## K-Nearest Neighbors Model

```python
def k_nearest_neighbors():
    
    #Create a model and print classification report
    KNN = KNeighborsClassifier()
    KNN.fit(X_train,y_train)
    predictions = KNN.predict(X_test)
    print("K-Nearest Neighbors Model Classification report")
    print(classification_report(y_test, predictions))
    
    # plot Confusion matrix
    cm_KNN_test = confusion_matrix(y_test, predictions, labels = ['b','g'])
    df_KNN_test = pd.DataFrame(data=cm_KNN_test, columns=['Predicted b', 'Predicted g'], index=['Actual b', 'Actual g'])
    sns.heatmap(df_KNN_test, annot=True)
    plt.show()
    
def k_nearest_neighbors_learningcurve():
    #K-Nearest Neighbors learning curve for overfitting and underfitting analysis
    pline=make_pipeline(StandardScaler(),KNeighborsClassifier())
    model=pline.fit(X,y)
    train_sizes, train_scores, test_scores=learning_curve(model, X, y,cv=5,train_sizes=np.linspace(.1, 1.0, 5),shuffle=True)
    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)
    plt.plot(train_sizes,train_mean,label="Training Accuracy")
    plt.plot(train_sizes,test_mean,label="Validation Accuracy")
    plt.title("Learning Curve")
    plt.xlabel("Number of data points")
    plt.ylabel("Accuracy")
    plt.legend(loc="best")
    
def k_nearest_neighbors_hyperparameter_tuning():
    pLine= Pipeline([('scale', StandardScaler()),('algo', KNeighborsClassifier())])
    params = {
        'algo__n_neighbors': np.arange(1,51,2),
        "algo__p" : [1,2],
        'algo__weights' : ['uniform', 'distance']
    }
    skf = StratifiedKFold(n_splits = 3)
    f1_scorer = make_scorer(f1_score, pos_label='g')
    knnGreadSearch = GridSearchCV(pLine, params, cv = skf, scoring = f1_scorer, verbose = 1)
    knnGreadSearch.fit(X_train,y_train)
    knnTuned = knnGreadSearch.best_estimator_
    y_test2 = knnTuned.predict(X_test)
    print("Report for KNN Model After HyperParameter Tuning")
    print(classification_report(y_test, y_test2))
    
def k_nearest_neighbors_auc_roc_curve():
    
    #K-Nearest Neighbors model Auc-Roc curve plotting
    pline = Pipeline([('scale', StandardScaler()),('algo', KNeighborsClassifier())])
    pline.fit(X_train, y_train)
    y_proba = pline.predict_proba(X_test)
    pred_1 = y_proba[:, 1]
    roc_curve(y_test, pred_1,pos_label = 'g')
    falsePR, truePR, threshold = roc_curve(y_test, pred_1, pos_label = 'g')
    aucScore = round((auc(falsePR, truePR)), 2)
    plt.figure(figsize=(10, 6))
    plt.plot(falsePR, truePR, 'g', label = f"AUC = {aucScore}")
    plt.plot([0, 1], [0, 1], 'r-.')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.xlabel('False Positive Rate ')
    plt.ylabel('True Positive Rate')
    plt.title('ROC-AUC curve')
    plt.legend()
    plt.show()
    
```

## K-Nearest Neighbors Model Results

```python
k_nearest_neighbors()
k_nearest_neighbors_learningcurve()
k_nearest_neighbors_hyperparameter_tuning()
k_nearest_neighbors_auc_roc_curve()
```

## Support Vector Machine Model

```python
def support_vector_machine():
    
    #Create a model and print classification report
    Svc = SVC()
    Svc.fit(X_train,y_train)
    predictions = Svc.predict(X_test)
    print("Support Vector Machine Model Classification Report")
    print(classification_report(y_test, predictions))
    
    # plot Confusion matrix
    cm_Svc_test = confusion_matrix(y_test, predictions, labels = ['b','g'])
    df_Svc_test = pd.DataFrame(data=cm_Svc_test, columns=['Predicted b', 'Predicted g'], index=['Actual b', 'Actual g'])
    sns.heatmap(df_Svc_test, annot=True)
    plt.show()
    
def support_vector_machine_learningcurve():
    #SVM learning curve for overfitting and underfitting analysis
    pline=make_pipeline(StandardScaler(),SVC())
    model=pline.fit(X,y)
    train_sizes, train_scores, test_scores=learning_curve(model, X, y,cv=5,train_sizes=np.linspace(.1, 1.0, 5),shuffle=True)
    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)
    plt.plot(train_sizes,train_mean,label="Training Accuracy")
    plt.plot(train_sizes,test_mean,label="Validation Accuracy")
    plt.title("Learning Curve")
    plt.xlabel("Number of data points")
    plt.ylabel("Accuracy")
    plt.legend(loc="best")

def support_vector_machine_auc_roc_curve():
    
    #SVM model Auc-Roc curve plotting
    pline = Pipeline([('scale', StandardScaler()),('algo', SVC(probability=True))])
    pline.fit(X_train, y_train)
    y_proba = pline.predict_proba(X_test)
    pred_1 = y_proba[:, 1]
    roc_curve(y_test, pred_1,pos_label = 'g')
    falsePR, truePR, threshold = roc_curve(y_test, pred_1, pos_label = 'g')
    aucScore = round((auc(falsePR, truePR)), 2)
    plt.figure(figsize=(10, 6))
    plt.plot(falsePR, truePR, 'g', label = f"AUC = {aucScore}")
    plt.plot([0, 1], [0, 1], 'r-.')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.xlabel('False Positive Rate ')
    plt.ylabel('True Positive Rate')
    plt.title('ROC-AUC curve')
    plt.legend()
    plt.show()
```
